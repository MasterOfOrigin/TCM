```python
import pandas as pd
import numpy as np
```

```python
param_names = ['R*', 'f_p', 'n_e','f_l', 'f_i',  "L'"]

r_dist = [
    [1,0.2],
    [5,0.4],
    [10,0.3],
    [20,0.1]
]
df_r = pd.DataFrame(r_dist, columns=['R*', 'p'])

fp_dist = [
    [0.7,0.2],
    [0.9,0.3],
    [1,0.5]
]
df_fp = pd.DataFrame(fp_dist, columns=['f_p', 'p'])


n_e_dist = [
    [0.25,0.2],
    [0.5,0.5],
    [1,0.2],
    [2,0.1]
]
df_n_e = pd.DataFrame(n_e_dist, columns=['n_e', 'p'])


f_l_dist = [
    [10**(-9),0.1],
    [10**(-6),0.1],
    [10**(-3),0.15],
    [0.1,0.15],
    [0.5,0.3],
    [0.9,0.2],
]
df_f_l = pd.DataFrame(f_l_dist, columns=['f_l', 'p'])


f_i_dist = [
    [10**(-5),0.1],
    [10**(-4),0.15],
    [10**(-3),0.3],
    [0.01,0.3],
    [0.05,0.15],
]
df_f_i = pd.DataFrame(f_i_dist, columns=['f_i', 'p'])


L_dist = [
    [5*10**(4),0.25],
    [10**(6),0.25],
    [5*10**(7),0.25],
    [10**9,0.25],
]
df_L = pd.DataFrame(L_dist, columns=["L'", 'p'])

df_cross = df_r.merge(
        df_fp, how='cross'
    ).merge(
        df_n_e, how='cross'
    ).merge(
        df_f_l, how='cross'
    ).merge(
        df_f_i, how='cross'
    ).merge(
        df_L, how='cross'
    )

table=df_cross[param_names].merge(
    df_cross.drop(param_names, axis=1).product(1).rename('p'),
    left_index=True,right_index=True
).sort_values('p', ascending=False).assign(
    n_planets=lambda x: x.drop('p', axis=1).product(1)
).assign(
    f_c_90=lambda x: -1*(0.1**(1/x['n_planets'])-1),
    f_c_95=lambda x: -1*(0.05**(1/x['n_planets'])-1),
    f_c_99=lambda x: -1*(0.01**(1/x['n_planets'])-1),
).assign(
    f_c_95_weighted=lambda x: x['f_c_95']*x['p']
)
```

```python
f_c_95_pcum = table.sort_values('f_c_95').assign(
    p_cum=lambda x: x['p'].cumsum()
)[['f_c_95', 'p_cum']]

f_c_95_pcum = f_c_95_pcum[(f_c_95_pcum['f_c_95'] < 1)]
```

```python
import seaborn as sns


ax = sns.displot(data=f_c_95_pcum, x='f_c_95', y='p_cum', log_scale=True)
ax.set(xlabel="f_c'", ylabel="Probability of Fermi Paradox", title="PFP(f_c') > 0.95")
```

```python
fc_95_pcum_table = []
for i in range(9):
    fc_95_pcum_table.append([10**(-i), table[table['f_c_95'] < 10**(-i)]['p'].sum()])
    
fc_95_pcum_table = pd.DataFrame(fc_95_pcum_table, columns=["f_c'", 'Pr Fermi (>0.95)']).sort_values("f_c'").style
fc_95_pcum_table
```
